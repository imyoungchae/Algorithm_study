```python
#     

import tensorflow as tf

height = 167
shoe_Size = 235


a = tf.Variable(0.1)
b = tf.Variable(0.2)

#
#return==(-)^2

#=*a+b
def los_function():
    predicted_value = height*a+b
    return tf.square(235 - predicted_value)

opt = tf.keras.optimizers.Adam(learning_rate=0.1) #  

for i in range(300):
    opt.minimize(los_function,var_list=[a,b]) # (, var_list=[a,b])
    #var_list :   
    print(height*a.numpy()+b.numpy())
```
