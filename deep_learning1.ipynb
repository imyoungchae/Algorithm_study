```python
# 내 키로 신발 사이즈를 추론해보기

import tensorflow as tf

'''키 = [170,182,167,162,178]
신발 = [270,275,235,230,275]'''

height = 167
shoe_Size = 235
#신발사이즈 = 키 * a + b

a = tf.Variable(0.1)
b = tf.Variable(0.2)

#손실함수'
#return=손실값=(실제값-예측값)^2
#예측값=키*a+b
def los_function():
    predicted_value = height*a+b
    return tf.square(235 - predicted_value)

opt = tf.keras.optimizers.Adam(learning_rate=0.1) # 경사하강법 도와준다

for i in range(300):
    opt.minimize(los_function,var_list=[a,b]) # (손실함수, var_list=[a,b])
    #var_list : 경사하강법으로 업데이트할 목록
    print(height*a.numpy()+b.numpy())
```
