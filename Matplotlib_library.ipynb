Matplotlib 라이브러리
- 다양한 데이터를 시각화 할 수 있도록 도와주는 라이브러리

##간단한 직선 그래프

```python
import matplotlib.pyplot as plt

x=[1,2,3]
y=[2,3,4]
plt.plot(x,y)
plt.title('My Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

##삼각함수 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,np.pi*10,500)
fig,axes=plt.subplots(2,1)
axes[0].plot(x,np.sin(x))
axes[1].plot(x,np.cos(x))
fig.savefig("sin&cos.png")
```

##이차함수 선 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-9,10)
y=x**2
plt.plot(x,y,linestyle=':',marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-9,10)
y1=x**2
y2=-2*x
plt.plot(x,y1,linestyle=':',marker='*',color='red',label='y=x*x')
plt.plot(x,y2,linestyle=':',marker='.',color='green',label='y=-2*x')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(
    shadow=True,
    borderpad=1
)
plt.show()
```
