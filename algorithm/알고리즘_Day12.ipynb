#스택
- 한 쪽 끝에서만 항목을 삭제하거나 새로운 item 저장
- 새로운 item 저장하는 연산: push
- Top item을 삭제하는 연산: pop
- LIFO(Last In First Out)


image.png

```python
def push(item):
    stack.append(item)

def peek():
    if len(stack) != 0:
        return stack[-1]
def pop():
    if len(stack) != 0:
        item=stack.pop(-1)
        return item

stack=[]
push('apple')
push('orange')
push('cherry')
print("사과, 오렌지, 체리 push 후: \t",end='')
print(stack,'\t<- top')
print('-top 항목: ',end='')
print(peek())

push('pear')
print('배 push 후 : \t\t',end='')
print(stack,'\t<- top')
print('-top 항목: ',end='')
print(peek())

pop()
push('grape')
print('pop(), 포도 push 후: \t',end='')
print(stack,'\t<-top')
print('-top 항목: ',end='')
print(peek())
```

- 연결리스트로 구현한 스택

```python
class Node:
    def __init__(self,item,link):
        self.item=item
        self.next=link

def push(item):
    global top
    global size
    top=Node(item,top)
    size += 1

def peek():
    if size != 0:
        return top.item

def pop():
    global top
    global size
    if size != 0:
        top_item=top.item
        top=top.next
        size -= 1
        return top_item

def print_stack():
    print('top->\t',end='')
    p=top
    while p:
        if p.next != None:
            print(p.item,'->',end='')
        else:
            print(p.item,end='')
        p=p.next
    print()
top=None
size=0

push('apple')
push('orange')
push('cherry')
print("사과, 오렌지, 체리 push 후: \t",end='')
print_stack()
print('-top 항목: ',end='')
print(peek())

push('pear')
print('배 push 후 : \t\t',end='')
print_stack()
print('-top 항목: ',end='')
print(peek())

pop()
push('grape')
print('pop(), 포도 push 후: \t',end='')
print_stack()
print('-top 항목: ',end='')
print(peek())
```

```python
def is_palindrome(input_string):
    token_list = list(input_string)
    flag=True
    while len(token_list) > 1 and flag == True:
        first_char = token_list.pop(0)
        last_char = token_list.pop(-1)
        if first_char != last_char:
            flag = False
    return flag

print(is_palindrome('racecar'))
print(is_palindrome('raddar'))
print(is_palindrome('hey'))
```

#큐
- 삽입과 삭제가 양 끝에서 각각 수행되는 자료구조
- FIFO(First In First Out) 원칙 하에 item의 삽입과삭제 수행
- front: 시작 / rear: 끝

```python
def add(item):
    q.append(item)

def remove():
    if len(q) != 0:
        item = q.pop(0) # 스택과 다르게 pop(0)이라고 표시 해주어야 함
        return item

def print_q():
    print('front -> ',end='')
    for i in range(len(q)):
        print('{!s:<8}'.format(q[i]),end='')
    print('<- rear')

q=[]
add('apple')
add('orange')
add('cherry')
add('pear')
print('사과, 오렌지, 체리, 배 삽입 후: \t', end='')
print_q()
remove()
print('remove한 후: \t\t',end='')
print_q()
remove()
print('remove한 후: \t\t',end='')
print_q()
```

- 연결리스트로 구현한 큐

```python
class Node:
    def __init__(self,item,n):
        self.item=item
        self.next=n

def add(item):
    global size
    global front
    global rear
    new_node = Node(item,None)
    if size == 0:
        front = new_node
    else:
        rear.next = new_node
    rear = new_node
    size += 1

def remove():
    global size
    global front
    global rear
    if size != 0:
        fitem = front.item
        front = front.next
        size -= 1
        if size == 0:
            rear = None
        return fitem

def print_q():
    p=front
    print('front: ',end='')
    while p:
        if p.next != None:
            print(p.item,'->',end='')
        else:
            print(p.item,end='')
        p=p.next
    print(' : rear')
front = None
rear =  None
size = 0
add('apple')
add('orange')
add('cherry')
add('pear')
print('사과, 오렌지, 체리, 배 삽입 후: \t', end='')
print_q()
remove()
print('remove한 후: \t\t',end='')
print_q()
remove()
print('remove한 후: \t\t',end='')
print_q()
```

```python
from collections import deque
dq = deque('data')
for elem in dq:
    print(elem.upper(),end='')

print()
dq.append('r')
dq.appendleft('k')
print(dq)

dq.pop()
dq.popleft()
print(dq[-1])
print('x' in dq)

dq.extend('structure')
dq.extendleft(reversed('python'))
print(dq)
```

#트리
- 미리 정렬해놓으면 이진탐색을 통해 효율적인 탐색 가능하지만
삽입이나 삭제 후에도 정렬상태를 유지해야하므로 삽입이나 삭제에 O(N)시간 소요
- 트리는 이진트리와 일반적인 트리로 구분
- 트리는 empty이거나 empty가 아니면 루트노드 R과 각 트리의 집합으로 구성되는데 각 트리의 루트노드는 R의 자식노드이다. 
단, 트리의 집합은 공집합일 수도 있다

*트리의 용어*

*루트(Root): 트리의 최상위에 있는 노드*

*자식(Child): 노드하위에 연결된 노드*

*차수(Degree): 자식 노드 수*

*부모(Parent): 노드의 상위에 연결된 노드*

*이파리(Leaf): 자식이 없는 노드*

*조상(Ancestor): 루트까지의 경로상에 있는 모든 노드들의 집합*

*후손(Descendant): 노드 아래로 메달린 모든 노드들의 집합*

*서브트리(Subtree): 노드 자신과 후손노드로 구성된 트리*

*레벨: 루트는 레벨1, 아래 층으로 내려가며 레벨이 1씩 증가*
*레벨은 깊이와 동일*

*높이: 트리의 최대 레벨*

*키: 탐색에 사용되는 노드에 저장된 정보 이탤릭체 텍스트 *

**N개의 노드가 있는 최대 차수가 k인 트리**

None 레퍼런스 수: Nk-(N-1)=N(k-1)+1

k가 클수록 메모리의 낭비가 심해지는 것은 물론 트리를 탐색하는 과정에서 None 레퍼런스를 확인해야 하므로 시간적으로도 매우 비효율적
